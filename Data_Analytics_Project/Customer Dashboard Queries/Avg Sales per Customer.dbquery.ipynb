```sql
-- Average sales per customer 
WITH CustomerSales AS (
    SELECT
        dim_customers.customer_id,
        SUM(fact_sales.price * fact_sales.quantity) AS TotalSales
    FROM
        fact_sales
    JOIN
        dim_customers ON fact_sales.customer_key = dim_customers.customer_key
    GROUP BY
        dim_customers.customer_id
)
SELECT
    AVG(TotalSales) AS AverageSalesPerCustomer
FROM
    CustomerSales;
```
