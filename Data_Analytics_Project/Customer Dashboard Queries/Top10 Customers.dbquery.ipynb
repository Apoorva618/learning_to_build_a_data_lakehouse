```sql
WITH ranked_customers AS (
  SELECT
    c.`customer_key`,
    c.`customer_id`,
    c.`customer_number`,
    c.`first_name`,
    c.`last_name`,
    SUM(s.`sales_amount`) AS total_sales,
    RANK() OVER (ORDER BY SUM(s.`sales_amount`) DESC) AS sales_rank
  FROM
    `workspace`.`salesdb`.`fact_sales` s
      JOIN `workspace`.`salesdb`.`dim_customers` c
        ON s.`customer_key` = c.`customer_key`
  WHERE
    s.`sales_amount` IS NOT NULL
    AND s.`customer_key` IS NOT NULL
  GROUP BY
    c.`customer_key`,
    c.`customer_id`,
    c.`customer_number`,
    c.`first_name`,
    c.`last_name`
)
SELECT
  *
FROM
  ranked_customers
WHERE
  sales_rank <= 10
ORDER BY
  sales_rank;
```
